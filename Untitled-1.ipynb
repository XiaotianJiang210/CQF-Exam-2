## 1.1 Background

### 1.1.1 Motivation and Practical Importance of Monte Carlo Pricing

In modern finance, **trading desks, risk managers, and structuring teams** routinely need to determine the fair value of derivatives whose payoffs depend on future asset prices. For plain-vanilla European options the **Black-Scholes Model** delivers a **closed-form** price. However, many real-world contracts (exotics, structured notes, embedded options, regulatory stress tests) do not admit analytical solutions. In those cases, practitioners turn to numerical methods.

Practitioners rely on **Monte Carlo (MC)** scheme to price complex instruments by simulating the uncertainty of future market movements. For example, instead of solving equations by hand, one can generate a large number of random price paths for an underlying asset and calculate the option's payoff for each path, the average discounted payoff, and get an estimate of the option's fair price. This simulation-based approach is especially valuable for exotic or **path-dependent** derivatives (e.g. Asian or barrier options), where payoff depends on the entire price path and no closed-form formula exists. In practice, banks and asset managers uses MC to value such derivatives and to perform risk analysis, because this technique can handle **any payoff structure and model complexity**. 

---

### 1.1.2 Risk-Neutral Pricing and Geometric Brownian Motion

Any discussion of option valuation in practice begins with **risk-neutral pricing**, the fundamental principle in modern derivatives theory. Under the risk-neutral measure (also known as the **Equivalent martingale measure**), **the current price of an asset equals the discounted expectation of its future price**. In other words, if markets have no arbitrage opportunities, a derivative's fair value can be computed as the present value of its expected payoff under risk-neutral probabilities. 

This concept is crucial for MC simulations: it tells us how to simulate the underlying asset's behavior for pricing purposes. We assume that investors are indifferent to risk (risk-neutral), which implies the drift of the asset price can be taken as the risk-free rate. By doing so, **the expected growth of the asset is neutralized to risk-free rate**, and discounting the payoff at the risk-free rate yields the correct no-arbitrage price. 

To model the evolution of the underlying asset's price under the risk-neutral measure, we typically use Geometric Brownian motion (GBM) as a baseline. GBM is the stochastic process assumed by the Black-Scholes model, capturing the idea that asset prices follow **continuous random walk with drift and volatility**. In a GBM model, the logarithmic returns of the asset are normally distributed, which reflects realistic features like proportional prices changes. Practically, GBM is defined by a stochastic differential equation (SDE): 
$$
\mathrm{d}S_t = r\,S_t\,\mathrm{d}t + \sigma\,S_t\,\mathrm{d}W_t
$$  
- **$S_t$**: Asset price at time $t$  
- **$\mu$**: Drift (expected rate of return)  
- **$\sigma$**: Volatility (diffusion coefficient)  
- **$dW_t$**: Increment of a Wiener process (standard Brownian motion)

GBM is appeal for practitioners is that it's both theoretically tractable (leading to the BSM formula) and intuitive, it ensures prices stay positive and incorporates compounding returns. It provides a reasonable first approximatin for many assets price dynamics. Even when more complex processes are needed (stochastic volatility, jumps, etc.), the risk-neutral pricing framework remains the same: simulate under the risk-free drift and discount the expected payoff. 

---

### 1.1.3 Simulating Price Paths: Euler - Maruyama and Milstein Schemes

While GBM has a known analytical solution for asset prices **(a lognormal distribution)**, in a simulation project we often discretize the time evolution to generate sample paths. Monte Carlo pricing requires constructing many possible paths of **$S_t$** from today to option expiry. To do this on a computer, one must solve the SDE numerically. Two commonly used numerical schemes for SDE simulation are the **Euler-Maruyama method and Milstein method**. 

#### - The Euler-Maruyama Scheme

The Euler-Maruyama scheme is a **simple and widely-used method for numerically approximating solutions to stochastic differential equations**. By discretizing the time interval, this scheme allows us to simulate GBM paths step by step. At each step, we update the GBM path by adding a Wiener increment, which is directly related to the square root of the time step. However, the Euler-Maruyama scheme introduces **some biases and may produce significant numerical errors when dealing with large time steps or volatile processes**, which will be revealed in the latter part of this project. The mathematical expression of the Euler-Maruyama scheme for a stock price increment is: 
$$
S_{t+\delta t}=S_{t}\Bigl(1+\mu\,\delta t+\sigma\sqrt{\delta t}\,\phi\Bigr)
$$

#### - The Milstein Scheme

To address the biases inherent in the Euler-Maruyama scheme, we introduce the Milstein scheme. This scheme offers an improvement by incorporating a **correction term** that takes into account the second derivative of the SDE. This correction reduces the bias and enhances the accuracy of the approximation. **By including this additional correction, the Milstein scheme performs better than Euler-Maruyama scheme**, especially when dealing with highly volatile processes or non-linearities. The mathematical expression of the Milstein scheme for a stock price increment is: 
$$
S_{t+\delta t}=S_{t}\Bigl(1+(\mu-\tfrac12\sigma^{2})\,\delta t
      +\sigma\sqrt{\delta t}\,\phi
      +\tfrac12\sigma^{2}\,\phi^{2}\,\delta t\Bigr)
$$

When it comes to simulating Geometric Brownian Motion, both the Euler-Maruyama and Milstein schemes have their strengths and limitations. The Euler-Maruyama scheme is straightforward and efficient, making it suitable for quick simulations. However, its accuracy might suffer in certain scenarios, **particularly when dealing with large time steps or high volatility.**

On the other hand, the Milstein scheme offers improved accuracy and can handle processes with non-linearities or significant volatility more effectively. While it adds some computational complexity compared to the Euler-Maruyama scheme, **the Milstein scheme remains widely used in financial modeling due to its superior ability to capture complex dynamics accurately.**

### 1.1.4 Project Objectives

This project addresses the core valuation problem: estimating  

$$
V_0 \;=\; e^{-r(T-t)}\,\mathbb{E}^{\mathbb{Q}}\!\bigl[\text{Payoff}(S_T)\bigr],
\qquad
\text{with}\;
\mathrm{d}S_t = r\,S_t\,\mathrm{d}t + \sigma\,S_t\,\mathrm{d}W_t .
$$  

Two benchmark instruments are studied:

| Contract | Pay-off at expiry $T$ | Practical relevance |
|----------|-----------------------|---------------------|
| **European call** | $\max(S_T-K,\,0)$ | Building block for delta-hedging, IV surfaces |
| **Binary (cash-or-nothing) call** | $\mathbf{1}_{\{S_T>K\}}$ | Common in structured notes; tied to option Greeks |

Pricing is performed via three complementary routes:

| Route | Purpose | Core idea |
|-------|---------|-----------|
| **Closed-form benchmark** | GBM closed form | $S_t \;=\;S_0\,\exp\!\bigl(\bigl(r - \tfrac12 \sigma^2\bigr)\,t \;+\;\sigma\,W_t\bigr)$ |
| **Euler–Maruyama MC** | Baseline discretisation | $S_{n+1}=S_n+rS_n\Delta t+\sigma S_n\sqrt{\Delta t}\,Z_n$ |
| **Milstein MC** | Higher-order accuracy | Adds $\tfrac12\sigma^{2}S_n\!\bigl((\sqrt{\Delta t}\,Z_n)^{2}-\Delta t\bigr)$ |

**Simulation workflow**

1. **Path generation** – simulate $N$ paths on a grid of $M$ steps ($\Delta t=T/M$).  
2. **Pay-off evaluation** – compute the option payoff at $T$ for each path.  
3. **Discount & average**

   $$
   \widehat{V}_0 \;=\; e^{-r(T-t)}\,\frac{1}{N}\sum_{i=1}^{N} \text{Payoff}^{(i)} .
   $$  

4. **Diagnostics** – compare Monte-Carlo estimates with closed-form benchmarks; study convergence versus $N$ (statistical error) and $\Delta t$ (discretisation bias).

#### 1.1.5 Why Should Finance Professionals Care?

* **Market Reality – Complex Pay-offs:**  
  Exchange-traded calls and puts may enjoy closed-form prices, but the street’s *profit pool* is in exotics: autocallables, worst-of baskets, barrier notes, structured deposits, convertible bonds, embedded guarantees in insurance products, and regulatory capital hedges. None of these pay-offs admit analytic formulas.

* **Regulatory & Risk Pressures:**  
  Basel III / IV, FRTB, SA-CCR, SA-CVA and IFRS 17 require banks and insurers to generate thousands of risk-neutral scenarios each day to compute Expected-Shortfall, CVA, capital add-ons and fair-value reserves.  MC is the *only* method that scales to high-dimensional factor sets without an explosion of state space.

* **Model Validation & P&L Explain:**  
  Even for vanilla books, desks run MC as an independent control to reconcile P&L attribution, stress-test volatility surfaces and validate analytical greeks. A solid MC engine therefore focuses not just pricing, but *model governance,* audit trails and regulatory self-assessment.

### 2.1 Basic Setup and Black-Scholes Formula

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import time  # for optional CPU time measurements

# For consistent "randomness"
np.random.seed(42)

# Baseline parameters (feel free to adjust)
S0    = 100.0   # Initial stock price
K     = 100.0   # Strike
r     = 0.05    # Risk-free rate
sigma = 0.20    # Volatility
T     = 1.0     # Time to maturity (in years)

# Functions for Black-Scholes analytics
def black_scholes_call(S0, K, r, sigma, T):
    """
    Closed-form Black-Scholes price for a European call.
    """
    if T <= 0:
        return max(S0 - K, 0)
    d1 = (np.log(S0 / K) + (r + 0.5*sigma**2)*T) / (sigma*np.sqrt(T))
    d2 = d1 - sigma*np.sqrt(T)
    call_price = S0*norm.cdf(d1) - K*np.exp(-r*T)*norm.cdf(d2)
    return call_price

def black_scholes_digital_call(S0, K, r, sigma, T):
    """
    Closed-form price for a digital (cash-or-nothing) call paying 1 if S_T > K.
    """
    if T <= 0:
        return 1.0 if S0 > K else 0.0
    d2 = (np.log(S0 / K) + (r - 0.5*sigma**2)*T) / (sigma*np.sqrt(T))
    return np.exp(-r*T)*norm.cdf(d2)
```

### 2.2 Monte Carlo Path Simulation

```python
def simulate_closed_form(S0, r, sigma, T, N, M=None):
    """
    Closed-form solution for simulating the underlying GBM:
    S_T = S0 * exp( (r - 0.5*sigma^2)*T + sigma * sqrt(T) * Z ).
    'M' is ignored here, but included for a consistent signature.
    """
    Z = np.random.randn(N)
    ST = S0 * np.exp((r - 0.5*sigma**2)*T + sigma*np.sqrt(T)*Z)
    return ST

def simulate_euler(S0, r, sigma, T, N, M):
    dt = T / M
    ST = np.full(N, S0, dtype=float)
    for _ in range(M):
        Z = np.random.randn(N)
        ST += r*ST*dt + sigma*ST*np.sqrt(dt)*Z
    return ST

def simulate_milstein(S0, r, sigma, T, N, M):
    dt = T / M
    # force float
    ST = np.full(N, S0, dtype=float)
    for _ in range(M):
        Z = np.random.randn(N)
        ST += (
            r*ST*dt 
            + sigma*ST*np.sqrt(dt)*Z
            + 0.5*sigma**2 * ST * dt * (Z**2 - 1.0)
        )
    return ST
```

### 2.3 Monte Carlo Pricing for European and Binary Calls

```python
def price_option_mc(simulator, S0, r, sigma, T, K, 
                    payoff_type="call", N=50_000, M=50):
    """
    simulator: one of the above (simulate_euler, simulate_milstein, simulate_closed_form)
    S0, r, sigma, T, K: the usual parameters
    payoff_type: "call" or "binary"
    N, M: #paths, #timesteps (M only matters for Euler/Milstein)
    """
    # 1) Generate terminal prices
    ST = simulator(S0, r, sigma, T, N, M)
    
    # 2) Payoff
    if payoff_type == "call":
        payoff = np.maximum(ST - K, 0.0)
    elif payoff_type == "binary":
        payoff = (ST > K).astype(float)
    else:
        raise ValueError("Unknown payoff type.")

    # 3) Discounted average
    disc_payoff = np.exp(-r*T)*payoff
    mean_est = np.mean(disc_payoff)
    std_est  = np.std(disc_payoff, ddof=1)
    std_err  = std_est / np.sqrt(N)
    return mean_est, std_err
```

### 2.4.1 Varying Number of Paths N

```python

S0    = 100.0
K     = 100.0
r     = 0.05
sigma = 0.20
T     = 1.0
np.random.seed(42)
import time

# For measuring times, we'll use time.time() around the calls
N_values = [10_000, 50_000, 100_000]
M = 50  # e.g. 50 time steps for Euler and Milstein

bs_call  = black_scholes_call(S0, K, r, sigma, T)
bs_digit = black_scholes_digital_call(S0, K, r, sigma, T)

print("Black–Scholes (reference) Prices:")
print(f"  European call = {bs_call:.4f}")
print(f"  Binary call   = {bs_digit:.4f}\n")

print("===  European Call  ===")
print(f"{'Method':<15s} {'N':>8s} {'Est. Price':>12s} {'Std. Err':>10s} {'CPU Time(s)':>12s} {'Diff vs BS':>10s}")
print("-"*70)

for N_ in N_values:
    # Euler
    t0 = time.time()
    price_eul, err_eul = price_option_mc(
        simulator   = simulate_euler,
        S0=S0, r=r, sigma=sigma, T=T, K=K,
        payoff_type = "call",
        N=N_,
        M=M
    )
    t1 = time.time()
    diff_eul = price_eul - bs_call
    
    # Milstein
    t2 = time.time()
    price_mil, err_mil = price_option_mc(
        simulator   = simulate_milstein,
        S0=S0, r=r, sigma=sigma, T=T, K=K,
        payoff_type = "call",
        N=N_,
        M=M
    )
    t3 = time.time()
    diff_mil = price_mil - bs_call
    
    # Closed-form sim
    t4 = time.time()
    price_cf, err_cf = price_option_mc(
        simulator   = simulate_closed_form,
        S0=S0, r=r, sigma=sigma, T=T, K=K,
        payoff_type = "call",
        N=N_,
        M=M
    )
    t5 = time.time()
    diff_cf = price_cf - bs_call
    
    print(f"Euler         {N_:8d} {price_eul:12.4f} {err_eul:10.4f} {(t1 - t0):12.2f} {diff_eul:10.4f}")
    print(f"Milstein      {N_:8d} {price_mil:12.4f} {err_mil:10.4f} {(t3 - t2):12.2f} {diff_mil:10.4f}")
    print(f"Closed-Form   {N_:8d} {price_cf:12.4f} {err_cf:10.4f} {(t5 - t4):12.2f} {diff_cf:10.4f}")
    print("")

print("===  Binary Call  ===")
print(f"{'Method':<15s} {'N':>8s} {'Est. Price':>12s} {'Std. Err':>10s} {'CPU Time(s)':>12s} {'Diff vs BS':>10s}")
print("-"*70)

for N_ in N_values:
    # Euler
    t0 = time.time()
    price_eul, err_eul = price_option_mc(
        simulator   = simulate_euler,
        S0=S0, r=r, sigma=sigma, T=T, K=K,
        payoff_type = "binary",
        N=N_,
        M=M
    )
    t1 = time.time()
    diff_eul = price_eul - bs_digit
    
    # Milstein
    t2 = time.time()
    price_mil, err_mil = price_option_mc(
        simulator   = simulate_milstein,
        S0=S0, r=r, sigma=sigma, T=T, K=K,
        payoff_type = "binary",
        N=N_,
        M=M
    )
    t3 = time.time()
    diff_mil = price_mil - bs_digit
    
    # Closed-form sim
    t4 = time.time()
    price_cf, err_cf = price_option_mc(
        simulator   = simulate_closed_form,
        S0=S0, r=r, sigma=sigma, T=T, K=K,
        payoff_type = "binary",
        N=N_,
        M=M
    )
    t5 = time.time()
    diff_cf = price_cf - bs_digit
    
    print(f"Euler         {N_:8d} {price_eul:12.4f} {err_eul:10.4f} {(t1 - t0):12.2f} {diff_eul:10.4f}")
    print(f"Milstein      {N_:8d} {price_mil:12.4f} {err_mil:10.4f} {(t3 - t2):12.2f} {diff_mil:10.4f}")
    print(f"Closed-Form   {N_:8d} {price_cf:12.4f} {err_cf:10.4f} {(t5 - t4):12.2f} {diff_cf:10.4f}")
    print("")
```

### 2.5 Sensitivity Results

#### 2.5.1 Helper Function for Parameter Sensitivity

```python
import numpy as np
import time

def sensitivity_sweep(simulator, param_name, param_values, payoff_type="call", 
                      N=50_000, M=50, 
                      S0=100.0, K=100.0, r=0.05, sigma=0.20, T=1.0):
    """
    Sweeps over 'param_values' for the parameter 'param_name'.
    For each value, runs Monte Carlo using 'simulator' and compares with B–S formula.
    
    simulator:   a function like simulate_euler(...) or simulate_milstein(...)
    param_name:  string in ["sigma", "r", "T", "S0"]
    param_values: a list of floats to test
    payoff_type: "call" or "binary"
    N:           number of Monte Carlo paths
    M:           number of time steps (for Euler/Milstein)
    S0, K, r, sigma, T: baseline parameters (overridden if param_name changes them)
    
    Returns a list of dicts with keys: 
      {
        'param_value': x,        # the tested parameter
        'MC_price': (float),
        'std_err':  (float),
        'BS_price': (float),
        'diff':     (MC_price - BS_price),
        'time':     (CPU time),
      }
    """
    
    base_S0    = S0
    base_r     = r
    base_sigma = sigma
    base_T     = T
    
    results = []
    
    for x in param_values:
        # Override the chosen parameter
        if param_name == "sigma":
            sigma = x
        elif param_name == "r":
            r = x
        elif param_name == "T":
            T = x
        elif param_name == "S0":
            S0 = x
        else:
            raise ValueError("Unsupported param_name in sensitivity_sweep.")
        
        # Measure CPU time for MC
        t0 = time.time()
        mc_price, std_err = price_option_mc(
            simulator   = simulator,
            S0          = S0,
            r           = r,
            sigma       = sigma,
            T           = T,
            K           = K,
            payoff_type = payoff_type,
            N           = N,
            M           = M
        )
        t1 = time.time()
        
        # Compute relevant Black–Scholes price
        if payoff_type == "call":
            bs_price = black_scholes_call(S0, K, r, sigma, T)
        else:
            bs_price = black_scholes_digital_call(S0, K, r, sigma, T)
        
        results.append({
            'param_value': x,
            'MC_price': mc_price,
            'std_err': std_err,
            'BS_price': bs_price,
            'diff': mc_price - bs_price,
            'time': (t1 - t0)
        })
        
        # Restore baseline
        sigma = base_sigma
        r     = base_r
        S0    = base_S0
        T     = base_T
        
    return results


def run_and_show_sensitivity(simulator, param_name, param_values, 
                             payoff_type="call",
                             N=50_000, M=50,
                             S0=100.0, K=100.0, r=0.05, sigma=0.20, T=1.0,
                             method_label="Euler"):
    """
    A small wrapper that calls 'sensitivity_sweep', then prints a table of results.
    """
    # Get the results
    data = sensitivity_sweep(simulator, param_name, param_values, payoff_type,
                             N, M, S0, K, r, sigma, T)
    
    # Print a header
    print(f"=== Sensitivity to {param_name} (Method={method_label}, payoff={payoff_type}) ===")
    print(f"{'Param':>8s}  {'MC_Price':>10s}  {'StdErr':>9s}  {'BS_Price':>10s}  {'Diff':>9s}  {'CPU(s)':>7s}")
    print("-"*60)
    
    # Print rows
    for row in data:
        print(f"{row['param_value']:8.3f}  "
              f"{row['MC_price']:10.4f}  "
              f"{row['std_err']:9.4f}  "
              f"{row['BS_price']:10.4f}  "
              f"{row['diff']:9.4f}  "
              f"{row['time']:7.3f}")
    print()

```

### 2.5.3 Sensitivity to Large Time Steps

```python
import numpy as np
import time

def time_steps_sweep(method, payoff_type="call", 
                     step_list=[1, 5, 12, 50, 100, 365],
                     N=50_000, 
                     S0=100, K=100, r=0.05, sigma=0.20, T=1.0):
    """
    For a given 'method' (simulate_euler / simulate_milstein / simulate_closed_form),
    vary the number of time steps in 'step_list' and compare the MC price with the B–S price.
    
    payoff_type: "call" or "binary"
    step_list: e.g. [1,5,12,50,100,365]
    N: # of MC paths
    S0, K, r, sigma, T: parameters for the option
    Returns a list of dicts: {M, MC_price, std_err, BS_price, diff, time}.
    """
    # Get the reference Black–Scholes price
    if payoff_type == "call":
        bs_price = black_scholes_call(S0, K, r, sigma, T)
    else:
        bs_price = black_scholes_digital_call(S0, K, r, sigma, T)
    
    results = []
    
    for M_ in step_list:
        t0 = time.time()
        # IMPORTANT: pass S0, r, sigma, T, K to match price_option_mc's signature
        mc_price, std_err = price_option_mc(
            simulator   = method,
            S0          = S0,
            r           = r,
            sigma       = sigma,
            T           = T,
            K           = K,
            payoff_type = payoff_type,
            N           = N,
            M           = M_
        )
        t1 = time.time()
        
        results.append({
            'M':        M_,
            'MC_price': mc_price,
            'std_err':  std_err,
            'BS_price': bs_price,
            'diff':     mc_price - bs_price,
            'time':     (t1 - t0)
        })
    return results
```

```python
S0    = 100.0
K     = 100.0
r     = 0.05
sigma = 0.40
T     = 1.0
N     = 50_000

step_list = [1, 5, 12, 50, 100, 365]

#do Euler
results_euler = time_steps_sweep(
    method       = simulate_euler,
    payoff_type  = "call",
    step_list    = step_list,
    N            = N,
    S0           = S0,
    K            = K,
    r            = r,
    sigma        = sigma,
    T            = T
)

# We can do Milstein similarly if we like
results_mil   = time_steps_sweep(
    method       = simulate_milstein,
    payoff_type  = "call",
    step_list    = step_list,
    N            = N,
    S0           = S0,
    K            = K,
    r            = r,
    sigma        = sigma,
    T            = T
)
```

```python
def convergence_data(simulator, payoff_type, Nmax=50000, batch=1000):
    """
    Generate a curve of MC estimates for every batch up to Nmax.
    Returns (paths_array_of_size_Nmax, partial_means_array).
    """
    # Step 1: Simulate all Nmax paths
    ST_all = simulator(S0, r, sigma, T, Nmax, M=50)
    
    # Step 2: Compute payoffs (European call example)
    if payoff_type == "call":
        payoffs_all = np.maximum(ST_all - K, 0.0) * np.exp(-r*T)
    else:
        payoffs_all = (ST_all > K).astype(float) * np.exp(-r*T)
        
    # Step 3: Calculate partial means in increments of 'batch'
    indices       = np.arange(batch, Nmax+1, batch)
    partial_means = []
    for n in indices:
        partial_means.append(np.mean(payoffs_all[:n]))
    
    return indices, partial_means

import matplotlib.pyplot as plt

Nmax = 50000
bs_price = black_scholes_call(S0, K, r, sigma, T)  # for call
indices, partial_means = convergence_data(simulate_euler, "call", Nmax, batch=1000)

plt.figure(figsize=(7,5))
plt.plot(indices, partial_means, label='Euler MC estimate')
plt.axhline(bs_price, color='red', linestyle='--', label='Black–Scholes Price')
plt.xlabel('Number of simulated paths (N)')
plt.ylabel('Option Price Estimate')
plt.title('Convergence of Monte Carlo (Euler) to B–S Price')
plt.legend()
plt.grid(True)
plt.show()
```

### 3.6 Challenge and Note on Random Seeds and Different Code Paths

One subtle source of differences in Monte Carlo estimates arose when comparing:

1. **Fixed runs** at a single volatility and interested based on certain number of paths.  
2. **Sweep runs** in which we loop over multiple parameter values (volatility ranges from 0.10, 0.20, 0.30, 0.40), calling the simulator for each one in turn.

Even if we set the same random seed (e.g., np.random.seed(42)) at the **start** of each script, once the code paths diverge, the **internal RNG state** evolves differently. For instance:

- In the **sweep** approach:  
  1. We first run MC at volatility = 0.1 → that consumes some large chunk of random draws (e.g., N * M normally distributed values).  
  2. Next, we do volatility = 0.2 → that consumes another chunk of random numbers.  
  3. Then volatility =0.3 … and so on.  

  By the time we finally get to volatility = 0.2 in that loop, **hundreds of thousands** of random draws have already been used for the volatility = 0.1 scenario.

- In the **fixed** approach:
  - We only run the simulator *once* at volatility = 0.2. It’s the very first call, so it grabs the initial block of random numbers from the generator without any prior consumption for other volatilities.

Hence, even though both scripts start with the same np.random.seed(...), the **order** of function calls (and how many random numbers each call consumes) changes the generator’s “pointer” in the pseudorandom sequence. Therefore, the set of Z = np.random.randn(...) values we get at volatility = 0.2 in the sweep code is **not** identical to the set we get in the “fixed volatility 0.2” code.  

- Both results are valid random samples from the same underlying distribution, just different draws.
- We typically see them differ by some fraction of the standard error, which is perfectly **normal** in Monte Carlo.

The mini‐project compared three methods for simulating the underlying stock price under geometric Brownian motion Euler Maruyama, Milstein, and the closed‐form lognormal approach—and used these paths in Monte Carlo to price two types of options:

Across multiple numerical experiments:
